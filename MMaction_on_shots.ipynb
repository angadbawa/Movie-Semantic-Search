```python
from mmaction.apis import inference_recognizer, init_recognizer
from mmengine import Config


# Choose to use a config and initialize the recognizer
config = 'configs/recognition/tsn/tsn_imagenet-pretrained-r50_8xb32-1x1x3-100e_kinetics400-rgb.py'
config = Config.fromfile(config)
# Setup a checkpoint file to load
checkpoint = 'checkpoints/tsn_r50_1x1x3_100e_kinetics400_rgb_20200614-e508be42.pth'
# Initialize the recognizer
model = init_recognizer(config, checkpoint, device = 'cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python
# Use the recognizer to do inference
from operator import itemgetter
video = '/content/drive/MyDrive/Movie semantic search/Shots/shot1040.mp4'
label = 'tools/data/kinetics/label_map_k400.txt'
results = inference_recognizer(model, video)

pred_scores = results.pred_score.tolist()
score_tuples = tuple(zip(range(len(pred_scores)), pred_scores))
score_sorted = sorted(score_tuples, key=itemgetter(1), reverse=True)
top5_label = score_sorted

labels = open(label).readlines()
labels = [x.strip() for x in labels]
results = [(labels[k[0]], k[1]) for k in top5_label]
actions=[]
for i in range(len(results)):
  if(results[i][1]>0.1887):
    actions.append(results[i])
print(actions)
```

```python
# Use the recognizer to do inference
from operator import itemgetter
import os
from tqdm import tqdm
folder_path = '/content/drive/MyDrive/Movie semantic search/Shots'

files = os.listdir(folder_path)

mp4_files = [file for file in files if file.endswith('.mp4')]
number_shots= len(mp4_files)
st = "shot"
mp = ".mp4"
shot = []
Actions = []
directory = '/content/drive/MyDrive/Movie semantic search/Shots/'
for i in tqdm(range(1,number_shots+1)):
  filename = f'shot{i}.mp4'
  filepath = os.path.join(directory, filename)
  video = filepath
  label = 'tools/data/kinetics/label_map_k400.txt'
  results = inference_recognizer(model, video)

  pred_scores = results.pred_score.tolist()
  score_tuples = tuple(zip(range(len(pred_scores)), pred_scores))
  score_sorted = sorted(score_tuples, key=itemgetter(1), reverse=True)
  top5_label = score_sorted[:10]

  labels = open(label).readlines()
  labels = [x.strip() for x in labels]
  Results = [(labels[k[0]], k[1]) for k in top5_label]
  actions=[]
  for i in range(len(Results)):
    if(Results[i][1]>0.1887):
      actions.append(Results[i])
  Actions.append(actions)
```

```python
print(len(Actions))
print(Actions)
action_list = []
for i in range(0, 401):
  if Actions[i] != []:
    temp_list = []
    for j in range(0, len(Actions[i])):
      temp_list.append(Actions[i][j][0])
    action_list.append(temp_list)
  else:
    action_list.append([])


print(action_list)
```

```python
import pandas as pd
df = pd.DataFrame()
df["Actions"] = action_list
df
```
